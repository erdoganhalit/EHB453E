```python

import pandas as pd
df = pd.read_excel(r'C:\Users\halit.erdogan\Downloads\itü-dataset_V3.xlsx', sheet_name='sipariş')
```

```python
saat_distribution = df['Siparis_Saati'].value_counts().to_dict()
saat_distribution
```

```python
import matplotlib.pyplot as plt
plt.bar(saat_distribution.keys(), saat_distribution.values(), width=0.6, color='green')
plt.xlabel("Hour of the day")
plt.ylabel("Number of samples")
plt.title("Distribution of Order Hours")
```

```python
restoran_distrib = df['RestaurantId'].value_counts().to_dict()
plt.bar(restoran_distrib.keys(), restoran_distrib.values(), width=0.6, color='green')
pos = range(len(restoran_distrib.keys()))
plt.xticks(pos, restoran_distrib.keys(), rotation=90)
plt.xlabel("Restaurant IDs")
plt.ylabel("Number of samples")
plt.title("Distribution of Restaurants")
```

```python
kurye_distrib = df['Kuryeid'].value_counts().to_dict()
kurye_distrib
```

```python
f = plt.figure()
f.set_figwidth(45)
f.set_figheight(20)

plt.bar(kurye_distrib.keys(), kurye_distrib.values(), width=0.6, color='green')
pos = range(len(kurye_distrib.keys()))
plt.xticks(pos, kurye_distrib.keys(), rotation=90)
plt.xlabel("Courier ID")
plt.ylabel("Number of samples")
plt.title("Distribution of Couriers")
```

```python
f = plt.figure()
f.set_figwidth(25)
f.set_figheight(20)

status_distrib = df['Statu'].value_counts().to_dict()
status_distrib

plt.bar(status_distrib.keys(), status_distrib.values(), width=0.6, color='green')
pos = range(len(status_distrib.keys()))
plt.xticks(pos, status_distrib.keys(), rotation=90)
plt.xlabel("Status Information")
plt.ylabel("Number of samples")
plt.title("Distribution of Status Information")
```

```python
f = plt.figure()
f.set_figwidth(25)
f.set_figheight(20)

status_distrib = df['Ata_Statu'].value_counts().to_dict()
status_distrib

plt.bar(status_distrib.keys(), status_distrib.values(), width=0.6, color='green')
pos = range(len(status_distrib.keys()))
plt.xticks(pos, status_distrib.keys(), rotation=90)
plt.xlabel("Ata Status Information", fontsize=20)
plt.ylabel("Number of samples", fontsize=20)
plt.title("Distribution of Ata Status Information", fontsize=20)
```
